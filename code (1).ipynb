```python

from ultralytics import YOLO
modelv8 = YOLO("yolov8n.pt")
from google.colab import drive
drive.mount('/content/drive')
import os
os.chdir('/content/drive/My Drive/yolo')
results = modelv8.train(data="data.yaml", epochs=10)
```

```python
#testing it with an image
resulttest2 = modelv8("/content/20240528_124206.jpg")
for result in resulttest2:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    result.show()  # display to screen
    result.plot()
    result.save(filename="result.jpg")  # save to disk
resulttest2[0].plot()
```

```python
#object counting
names= modelv8.names
syringe_id = list(names)[list(names.values()).index('pfs')]
vials_id = list(names)[list(names.values()).index('vials')]

syringe_count = resulttest2[0].boxes.cls.tolist().count(syringe_id)
vials_count = resulttest2[0].boxes.cls.tolist().count(vials_id)
```

```python
#printing the object count
print("No. of syringes in the image:",syringe_count,"\n  number of vials in the image are :",vials_count)
```
